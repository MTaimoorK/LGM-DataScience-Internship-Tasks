##### 1. __Importing Necessary Libraries and Dataset__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset
df = pd.read_csv("globalterrorismdb_0718dist.csv", encoding="latin1")
```

```python
data = pd.DataFrame(df, df.rename(columns={ 'success': 'Success', 'location': 'Location', 'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country',
                   'country': 'Country_code', 'provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                   'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                   'weaptype1_txt':'Weapon_type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'},inplace=True))
```

```python
data.head()
```

```python
#checking null values for specific attributes
data =  data[['Year','Month','Extended','Day','Country','State','Region','City','Latitude','Longitude','Location','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive', 'Success']]
```

#### 3. __Explaratory Analysis__

##### _City With Most Attacks_

```python
data.City.value_counts().to_frame().drop('Unknown').sort_values('City',axis=0,ascending=True).head(10).plot(kind='bar',figsize=(20,10),color='red')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Top 10 most affected cities",fontsize=20)
plt.show()
```

##### _Country With Most Attacks_

```python
plt.figure(figsize=(25,10))
sns.barplot(x=df['Country'].value_counts()[:5].index, y=data['Country'].value_counts()[:5].values, palette='rocket')
plt.title('Top 5 Most Dangerous Countries', fontsize=20)
plt.xlabel('Country', fontsize=15)
plt.ylabel('No. of attacks', fontsize=15)
plt.xticks(rotation=50)
plt.show()
```

#### _Regions With Most Attacks_

```python
data['Region'].value_counts().to_frame().sort_values('Region',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(25,10),color='red')
plt.xticks(rotation = 50)
plt.xlabel("Region",fontsize=15)
plt.ylabel("No. of attacks",fontsize=15)
plt.title("Top 10 Most Dangerous Regions",fontsize=20)
plt.show()
```

#### _Most Targeted Types_

```python
plt.subplots(figsize=(25,10))
sns.countplot(data=data, x=data['Target_type'],order=data['Target_type'].value_counts().index,palette="rocket")
plt.xticks(rotation=90)
plt.xlabel("Target Type",fontsize=15)
plt.ylabel("No. of attacks",fontsize=15)
plt.title("Most Targeted Types",fontsize=20)
plt.show()
```

##### _Most Used Attack Types_

```python
data['AttackType'].value_counts().plot(kind='bar',figsize=(20,10),color='red')
plt.xticks(rotation = 50)
plt.xlabel("Attack Type",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Most used attack types",fontsize=20)
plt.show()
```

##### _Highest Number of Attacks by Groups_

```python
data[['Group','Killed']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Killed',ascending=False).head(10).plot(kind='bar',color='red',figsize=(25,10))
plt.title("Top 10 terrorist attack groups",fontsize=20)
plt.xlabel("Terrorist group",fontsize=20)
plt.ylabel("Number of attacks",fontsize=20)
plt.show()
```
